```python
import torch
from transformers import AutoModelForCausalLM, BitsAndBytesConfig

model_path = "/content/model-fp16"

bnb_config = BitsAndBytesConfig(
    load_in_8bit=True
)

model_int8 = AutoModelForCausalLM.from_pretrained(
    model_path,
    quantization_config=bnb_config,
    device_map="auto"
)

model_int8.save_pretrained("/content/quantized/model-int8")

print("✅ INT8 DONE")
```

```python
import torch
from transformers import AutoModelForCausalLM, BitsAndBytesConfig

model_path = "/content/model-fp16"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4"   # 🔥 best quality 4-bit
)

model_int4 = AutoModelForCausalLM.from_pretrained(
    model_path,
    quantization_config=bnb_config,
    device_map="auto"
)

model_int4.save_pretrained("/content/quantized/model-int4")

print("✅ INT4 DONE")
```

```python
import os

def get_size(path):
    total = 0
    for root, dirs, files in os.walk(path):
        for f in files:
            total += os.path.getsize(os.path.join(root, f))
    return round(total / (1024**3), 2)

print("FP16:", get_size("/content/model-fp16"), "GB")
print("INT8:", get_size("/content/quantized/model-int8"), "GB")
print("INT4:", get_size("/content/quantized/model-int4"), "GB")
```

```python
import time
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("/content/model-fp16")

model = AutoModelForCausalLM.from_pretrained(
    "/content/quantized/model-int8",
    device_map="auto"
)

inputs = tokenizer("Explain machine learning", return_tensors="pt").to("cuda")

start = time.time()
_ = model.generate(**inputs, max_new_tokens=50)
end = time.time()

print("INT8 Speed:", round(end - start, 2), "sec")
```
